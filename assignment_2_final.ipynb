## STEP FUNCTION

The first perceptron was developed in 1943 by McCulloch & Pitts Â using a step activation function represented by the equation below.
           
           
 

The step function is defined by threshold.
If input value is greater than threshold, then output is 1, else 0.




```python
import matplotlib.pyplot as plt
import numpy as np

def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()

x = np.arange(-10, 11, 1)
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

## Advantages :

**1. It is very easy to understand and implement.**

**2. We can do a particularly useful binary classification task where it takes 1/0 as input and gives output as 1/0.**

**3. It can help to transform continuous inputs into discrete outputs.**

**4. We can use this function in a neural network as a nonlinear function.**

## Disadvantages :

**1. The step function produces a binary output (0 or 1) regardless of the input value. As a result, optimizing the neural network to get      reliable results is challenging.**

**2. The gradient information is lost in that range since the step function provides a constant output(1/0) for a range of input values.**

**3. This function is not 0-centered; many neural network topologies depend on the output values centered around 0. Therefore, it might be problematic for some of them.**

**4. The curve transition occurs quickly. With a huge dataset, it won't be easy to characterize.**

# SIGMOID FUNCTION

**The mathematical form of the sigmoid nonlinearity is shown below.**



**It compresses a real-valued number into the range of 0 and 1. As a result, it is particularly utilized for models whose output is a probability prediction.**





```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

> The Sigmoid Function curve looks like a S-shape.

## Advantages :

**1. It is suitable for use in binary classification problems at the output layer of neural networks since the sigmoid function's value lies between 0 and 1, and output may be predicted by maintaining a threshold value between 0 and 1.**

**2. It has a smooth gradient, preventing jumps in output values.**

## Disadvantages :

**1. Vanishing gradientâfor very high or very low values of X, the prediction has almost no change, causing a vanishing gradient problem. This can result in the network refusing to learn further or being too slow to reach an accurate prediction.**

**2. Its outputs are not zero centered**

**3. It is Computationally expensive**

# Tanh FUNCTION

**The hyperbolic tangent function is also a prevalent activation function for neural networks that must output values in the range between -1 and 1. This activation function is simply the hyperbolic tangent (tanh) function, as shown in the following equation:**






**Its activations saturate like those of the sigmoid neuron, but its output is zero-centered. As a result, in real-world applications, tanh nonlinearity always prevails over sigmoid nonlinearity. Also, keep in mind that the tanh neuron is just a scaled sigmoid neuron, and in particular, the following is true:**







```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

## Advantages :

**1. It is usually used in hidden layers of neural networks and gives values between -1 and 1. It will make mean for the hidden layer 0 or very close to itself and help in the centering of data and learning for the next layer much easier.**

**2. Effective for modeling extremely positive, neutral, and strongly negative inputs.**


## Disadvantages : 

**1. There is almost no change to the prediction for very high or very low values of x, causing a vanishing gradient problem. Due to this, a neural network may refuse to learn further, or it will become too slow to find the accurate output.**

**2. It is also computationally expensive.**

# Rectified Linear Units (RELU FUNCTION)

**The Rectified Linear Unit has become very popular in the last few years. The following equation shows the straightforward ReLU function:**



**The activation is thus simply thresholded at zero.**






```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

## Advantages:

**1. The Cost Of Computation Is Less As Compared To Sigmoid & Tan H Since It Uses Simple Mathematical Computations. This Function Activates Only A Few Neurons At A Time Which Makes The Neural Network Sparse & Computationally Efficient.**

**2. It Is Faster Than The Sigmoid & Tan H Activation Function.**

**3. The Derivative Function In ReLU Allows Backpropagation.**


## Disadvantages:

**1. The Network Cannot Perform Backpropagation When The Gradient Of The Function Becomes Zero Due To the Zero Or Negative Value Of The Input.**

# ELU FUNCTION

Its full form is Exponential Linear Units. To solve some of ReLU's drawbacks, including the "dying ReLU" issue, the ELU function was developed. 
it defines as : 







## Advantages :

**1. It has all the benefits of the ReLU function.**

**2. Closer to zero mean outputs**

**3. The fact that the ELU function aids in preventing the "dying ReLU" problem is one of its key advantages. When a ReLU neuron receives a negative input, it produces zero output, known as the "dying ReLU" issue. When this occurs, the neuron essentially "dies" and stops contributing to the neural network's output. The ELU function solves this issue by having a non-zero output for negative input values to keep the neurons active even when the input is negative.**


## Disadvantages :

**1. A negative saturation regime compared with Leaky ReLU adds some robustness to noise**

**2. The ELU function is more computationally expensive than ReLU**

**3. Not all varieties of neural networks can utilize the ELU function.**

# SELU FUNCTION 

**The Scaled Exponential Linear Unit (SELU) function is an activation function. The SELU function was devised to address some of the drawbacks of ELU, such as the demand for manual hyperparameter adjustment.**

**The SELU function is defined as:**

``f(x) = lambda * (max(0, x) + alpha * (exp(min(0, x)) - 1)),``

**Alpha and lambda are two variables automatically calculated during training from the input data. A smooth and continuous function is the SELU function.**


## Advantages :

**1. The SELU function's ability to self-normalize is one of its key benefits. This indicates that the function's output is nearly normally distributed, with a zero mean and a single variance, for a given input.**

**2. The gradients are kept from getting too tiny or too big by the function, which is built to maintain each neuron's output within a certain range.**

**3. Automatically computes the values of lambda and alpha during training**


## Disadvantages : 

**1. The SELU function is only applicable to feedforward neural networks with dense layers and may not work well for other types.**

**2. The SELU function is sensitive to weight initialization.**
